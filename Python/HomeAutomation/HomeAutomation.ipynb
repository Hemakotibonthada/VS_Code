```python
from flask import Flask, jsonify

app = Flask(__name__)

# Global variable to store the LED state
led_state = "OFF"

@app.route('/')
def index():
    return '''
    <h1>ESP32 Control</h1>
    <p>LED <a href="/on"><button>ON</button></a>&nbsp;<a href="/off"><button>OFF</button></a></p>
    '''

@app.route('/on', methods=['GET'])
def turn_on():
    global led_state
    if led_state != "ON":
        led_state = "ON"
    return jsonify({"status": f"LED is {led_state}"})

@app.route('/off', methods=['GET'])
def turn_off():
    global led_state
    if led_state != "OFF":
        led_state = "OFF"
    return jsonify({"status": f"LED is {led_state}"})

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
